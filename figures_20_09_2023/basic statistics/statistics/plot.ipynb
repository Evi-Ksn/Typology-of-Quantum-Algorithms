```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
plt.rcParams['axes.unicode_minus'] = False

fig, ax = plt.subplots(figsize=(15,4), dpi=100)

y_data = [ 48,  30, 28,17,13,13,6 ]
x_data = ('Dynamic system', 'Combinatorial','Linear algebra', 'Statistics','Hidden subgroup','Numerical optimization','Stochastic process')

# 柱状图颜色
colors = ['#0780cf','#765005','#fa6d1d','#0e2c82','#b6b51f','#da1f18','#38cb7d','#ddae33','#844bb3','#93c555','#5f6694','#df3881']

# 柱状图
bar = plt.bar(x_data, y_data, 0.5, color=colors,edgecolor='grey')

# 设置标题
ax.set_title('Fundamental math problems',fontsize=14,y=1.05)
# 设置坐标轴标题
ax.set_ylabel("",fontsize = 12,color = 'black',alpha = 0.7,rotation=360)
# 设置Y轴区间
ax.set_ylim(0,55)

# 显示数据标签
for a,b in zip(x_data, y_data):
    plt.text(a,b,
             b,
             ha='center',
             va='bottom',
            )
plt.savefig('Fundamental math problems.png',dpi=300,bbox_inches='tight')
# 边框隐藏
# ax.spines['top'].set_visible(False)
# ax.spines['right'].set_visible(False)
# ax.spines['bottom'].set_visible(False)
# ax.spines['left'].set_visible(False)
```
